Lagrangiana de um pêndulo simples

```python
# -*- coding: utf-8 -*-
import numpy as np
from numpy import sin, cos
from scipy.integrate import odeint
from matplotlib import pyplot as plt
from jupyterthemes import jtplot
```

```python
def equacoes(x0,t):
    theta, x = x0
    f = [x, -(g/l)*sin(theta)]
    return f

def plot_results(tempo, theta_1, theta_2):
    plt.plot(tempo, theta_1[:,0])
    plt.plot(tempo, theta_2)
    
    s = ("ângulo incial= " + str(angulo_inicial) + " graus")
    plt.title('movimento do pêndulo: ' + s)
    plt.xlabel('tempo (s)')
    plt.ylabel('ângulo (rad)')
    plt.grid(True)
    plt.legend(['não linear', 'linear'], loc= 'lower left')
    plt.figure()
    jtplot.style(theme='grade3', ticks=True, grid=True, figsize=(20.5, 15.5), spines=True)
    plt.show()
```

```python
#parametros
g = 9.8 #gravidade[m/s]
l = 0.96 #comprimento [m]
tempo = np.arange(0,15.0, 0.020)

#condições iniciais
angulo_inicial = 32.8 #graus
massa = 0.09842 
theta0 = np.radians(angulo_inicial)
x0 = np.radians(0.0)

#achar a solução para o problema não linear
theta_1 = odeint(equacoes,[theta0,x0], tempo)

#solução para o problema linear
w = np.sqrt(g/l) #velocidade angular
theta_2 = [theta0 * cos(w*t) for t in tempo]

#plotar os resultados
plot_results(tempo, theta_1, theta_2)
```
